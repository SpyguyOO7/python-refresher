# Simulation

```python
import numpy as np
import physics
```

## Parameters

```python
density_water = 1000  # kg/m^3
density_object = 500  # kg/m^3
object_volume = 0.1  # m^3
g = 9.81  # m/s^2
time_step = .01  # s
time_max = 20  # s
```

```python
startTime = 0  # s
startDepth = 100  # m
startVelocity = -50  # m/s
#acceleration = 0 # m/s^2
buoyancy =  density_water*g* object_volume # N
weight =   density_object*object_volume*g# N
```

```python
# lists for plotting later
time = np.arange(0, time_max, time_step)
mass = density_object*object_volume
position = np.zeros_like(time)
positionExp = np.zeros_like(time)
velocity = np.zeros_like(time)
velocityExp = np.zeros_like(time)
acceleration = np.zeros_like(time)
zeroRef = np.zeros_like(time)
```

```python
# simulation loop
print(buoyancy)
print((buoyancy - weight)/mass)
currentVelo = startVelocity
currentPos = 0-startDepth
for i in range(1, len(time)):
    # calculate the force on the object

    force = buoyancy - weight
    x = i*time_step+startTime
    # calculate acceleration

    acceleration[i] = force/mass
    #if (x < 70):
    #    acceleration[i] = np.sin(x)*10+np.sin(2*x)*5+np.sin(7*x)*3+np.sin(6*x)*2+np.sin((1/6)*x)*2
    # calculate velocity
    velocityExp[i] = acceleration[i]* time_step + currentVelo 
    velocity[i] = (x* acceleration[i] ) + startVelocity
    # calculate position
    positionExp[i] = velocityExp[i]* time_step + currentPos 
    position[i]  = x*x*acceleration[i]*.5 + startVelocity*x - startDepth
    #exp stuff
    currentPos = positionExp[i]
    currentVelo = velocityExp[i]

    #if (abs(positionExp[i]-position[i])>1):
    #    position[i]
    #print (velocityExp[i])
```
